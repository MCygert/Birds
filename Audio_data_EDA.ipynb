```python
sound, sr = sf.read(path_to_file)
fig, ax = plt.subplots(figsize=(16, 8))
sns.lineplot(x=np.arange(len(sound)), y=sound, ax=ax);
```

```python
from scipy.fft import fft, fftfreq, ifft
# https://realpython.com/python-scipy-fft/#why-would-you-need-the-fourier-transform
def noise_reduction(y, sr, th=0.3):

    SAMPLE_RATE = 1
    DURATION = len(y) / SAMPLE_RATE
    N = int(SAMPLE_RATE * DURATION)

    yf = fft(y)
    xf = fftfreq(N, 1 / SAMPLE_RATE)
    new_yf = yf.copy()
    middle = len(y) / 2
    new_yf[int(middle - len(y) * th):int(middle + len(y) * th)] = 0
    new_y = ifft(new_yf)
    return new_y, sr, xf
```

```python
y, new_sr, x = noise_reduction(sound, sr)
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.lineplot(x=np.arange(len(y)), y=y, ax=ax);
```
